## 3. Pandas df operations for cleaning and/or pre-processing data

### 3.1. Sorting values

```python
df1.sort_values('Name', ascending=1).head()
```

```python
df1.sort_values(['Name','Type2', 'Speed'], ascending=[1,0,1]).head(20)  # only sort based on the first col
```

### 3.2. Operations between multiple columns and store into new column

```python
df2 = pd.DataFrame()
df2['total'] = df1.HP + df1.Defense
df2['total2'] = df1['Type1'] + '-' + df1['Type2'].apply(str) # addition with NaN will give NaN
df2['Attack'] = df1['Attack'] 
df2['Sp_atk'] = df1['Sp_atk']
df2['Speed'] = df1['Speed']
df2['total3'] = df2.iloc[:,2:5].sum(axis=1)
df2.head()
```

### 3.3. Rearranging column

```python
col2 = list(df2.columns)
neworder = [1,2,3,4,0,5]
col2 = [col2[i] for i in neworder]
df2 = df2[col2]
df2.head(10)
```

### 3.10. Use groupby to get stats aggregate (mean, median, count, max, min, sum, etc) or to analyze data

```python
df1.groupby('Type1').count()
```

```python
df1.groupby(['Type1', 'Type2']).mean() #--> will apply means on all columns when counts per group are >1
```

```python
df1.groupby(['Type1']).Legendary.sum() # --> will take sum of legendary columns for same group of Type1
```

```python
df1.groupby(['Type1', 'Type2']).mean().sort_values('HP', ascending=False)
```

```python
df1.groupby('Type1').size().plot(kind='bar')
```

```python
df1.groupby(df1.HP//25 *25).size()
# for grouping with numeric data, to use larger bin size, we can simply divide the numbers (ignore remainders with //)
# inside the groupby function for grouping, then multiply back with the divisor to get the right display numbers
```

```python
df1.groupby(['Type1',df1.HP//50*50]).size().plot(figsize=(15,15),kind='bar')   
# not recommended to directly plot nested groupby data.
```

```python
# unstacking multiple groups allow unpacking of data to plot along sensible dimensions
df1.groupby([df1.HP//50*50,'Type1']).size().unstack('HP').plot(figsize=(12,12), kind='bar', stacked='True', 
                                                            ylabel='Number of Pokemon')   
#df1.groupby([df1.HP//50*50,'Type1']).size().unstack('HP').fillna(0).plot(figsize=(12,12), kind='bar', stacked='True', 
#                                                            ylabel='Number of Pokemon')   
# fillna can be applied if some grouped vars gives NaN
```

```python
# Check grouped counts:
df1['Count']=1
print(df1[df1.Type1 == 'Flying'].Count.sum())
print(df1[(df1.Type1 == 'Bug') & (df1.HP<50)].Count.sum())
print(df1[(df1.Type1 == 'Water') & (df1.HP>150)].Count.sum())
```

```python
compare = df1[(df1.Type1 == 'Fire') | (df1.Type1 == 'Ice')].fillna(0).reset_index(drop=True).drop(['Unnamed: 0'], axis=1)
compare.head()
```

```python
compare.groupby(['Type1', compare.Attack//10*10]).size().unstack('Type1')
```

```python
compare.groupby(['Type1', compare.Attack//10*10]).size().unstack('Type1').fillna(0).plot()
```
